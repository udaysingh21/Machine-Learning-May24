```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Seaborn has some built-in datasets
```

```python
# Total Bill vs Tip Datasets
```

```python
# importing tips dataset into tips variable as pandas dataframe
tips=sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
sns.set_theme()
```

```python
# visualise the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset

iris=sns.load_dataset('iris')
iris.head()
```

```python
# Scatter Plot()

sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
# count plot
sns.countplot(x='class',data=titanic)
```

```python
titanic.shape
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
# Bar Plot
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
df=pd.read_csv('..\Pandas\diabetes.csv')
df.head()
```

```python
# Distribution Plot
```

```python
sns.displot(df['Glucose'])
```

```python
sns.distplot(df['Glucose'])
```

```python
# Co-relation Matrix
# positive and negative

corr=df.corr()
corr
```

```python
# Heat Map

plt.figure(figsize=(10,10))
sns.heatmap(corr,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
# color-dark : positive correlation
# light-color: negative correlation
```
