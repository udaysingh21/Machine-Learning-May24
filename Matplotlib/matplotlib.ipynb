```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100) # 100 equal space values between 0 and 10
y=np.sin(x) # x is the angle and y is sin(x)
z=np.cos(x)
```

```python
# Plotting the data

# sin wave
# plt.plot() # it will create empty figure

plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding titles and labels
plt.plot(x,y)
plt.xlabel('Angles')
plt.ylabel("Sine Values")
plt.title('Sine Wave')

plt.show()
```

```python
# Parabola
x=np.linspace(-10,10,20) 
# 20 equally spaced values between -10 and 10
y=x**2

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Parabola")

plt.show()
```

```python
plt.plot(x,y,'r+',)
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')

plt.xlabel('Angles')
plt.ylabel('Sin/Cos')
plt.legend(['Sin','Cos'])

plt.show()
```

```python
# Bar plot
fig=plt.figure() # empty plot
ax=fig.add_axes([0,0,1,1]) # ax is axis , area in which we want to have plot
languages=['c','c++','java','python','javascript']
popular=[10,15,25,40,10]

ax.bar(languages,popular)
plt.xlabel('Programming Languages')
plt.ylabel('Percentage')
plt.title('Bar Plot')

plt.show()
```

```python
# Pie Chart

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['c','c++','java','python','javascript']
popular=[10,15,25,40,10]

ax.pie(popular,labels=languages,autopct='%1.1f%%')
plt.xlabel('Programming Languages')
plt.ylabel('Percentage')
plt.title('Pie Chart')

plt.show()
```

```python
# Scatter Plot
```

```python
x=np.linspace(0,10,30) # 30 equally spaced values between 0 and 10

y=np.sin(x)
z=np.cos(x)

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# 3d Scatter Plot

fig=plt.figure()
ax=plt.axes(projection='3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)

# c is colors, c=x , this means that the colors of markers are determined by the values of x
ax.scatter(x,y,z,c=x,cmap='Blues')
plt.show()
```
